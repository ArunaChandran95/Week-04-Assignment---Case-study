```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Importing Dataset into the python environment:

```python
data = pd.read_csv("cars_data.csv")
```

```python
data.head()
```

# 2. Checking for any null values present in the dataset:

```python
data.isnull().sum()
```

```python
data.head()
df = pd.DataFrame(data)
```

# 3. Bar Graph- Buyers participated in the sales:

```python
data['Buyer Gender'].value_counts()
```

```python
data['Buyer Gender'].value_counts().plot(kind='bar');
plt.xlabel("Buyer Gender")
plt.ylabel("Total Count")
plt.title("Distribution of Male and Female buyers participated in the sales")
plt.show()

```

# 4. Top Five cars based on their sales price:


# 5. Least five cars based on their resell price:
